# KDE

```python
# This is called rejection sampling, a way to brute force sample any surface, so long
# as you're fine waiting for slow code
n = 50000
xs, ys = np.random.uniform(0, 10, n), np.random.uniform(0, 10, n)
zs = (np.sin(xs) + np.cos(ys) + 2 * np.arcsinh(xs * ys))**2
zs /= zs.max()
passed = np.random.uniform(0, 1, n) < zs
xs, ys = xs[passed], ys[passed]
plt.scatter(xs, ys, s=1, alpha=0.2);
```

# Pandas Styling for Tables

```python
df = pd.DataFrame(np.random.normal(size=(6,6)), columns=[x for x in "ABCDEF"])
df
```

```python
# create a function when the column value is less than 0, show red 
def neg_red(x):
    return f"color: {'red' if x < 0 else 'white'}"
df.style.applymap(neg_red)
```

```python
def gold_max(xs):
    m = xs.to_numpy().max()
    color = {True: "background-color: #c78f2e", False: ""}
    is_max = (xs == m).replace(color)
    return is_max
df.style.apply(gold_max, axis=None)
```

```python
df.style.background_gradient(cmap="magma")
```

# 4D Spiral

```python
np.random.seed(0)
n = 2000
t = np.linspace(0, 20, n)
x = t * np.sin(t) + 0.2 * np.random.normal(size=n)
y = t * np.cos(t) + 0.2 * np.random.normal(size=n)
z = np.log(t + 1) * np.sin(np.sqrt(t)) + 0.1 * np.random.normal(size=n)
a = np.log(t + 1) * np.cos(np.sqrt(t)) + 0.1 * np.random.normal(size=n)

df2 = pd.DataFrame({"x": x, "y": y, "z": z, "a": a})
data = df2.to_numpy()
```

```python
pd.plotting.scatter_matrix(df2);
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z, c=t, s=3+a);
```

# Manifold Learning

See https://scikit-learn.org/stable/modules/manifold.html 

```python
methods = [
    ("LLE", manifold.LocallyLinearEmbedding(n_neighbors=20, method='standard')),
    ("LTSA", manifold.LocallyLinearEmbedding(n_neighbors=20, method='ltsa')),
    ("Hessian LLE", manifold.LocallyLinearEmbedding(n_neighbors=20, method='hessian')),
    ("Modified LLE", manifold.LocallyLinearEmbedding(n_neighbors=20, method='modified')),
    ("Isomap", manifold.Isomap()),
    ("MDS", manifold.MDS(n_init=1)),
    ("SE", manifold.SpectralEmbedding()),
    ("t-SNE", manifold.TSNE()),
]

fig, axes = plt.subplots(nrows=2, ncols=4)
for (name, method), ax in zip(methods, axes.flatten()):
    print(f"Running {name}")
    Y = method.fit_transform(data)
    ax.scatter(Y[:, 0], Y[:, 1], c=t)
    ax.set_title(name)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4)
for (name, method), ax in zip(methods, axes.flatten()):
    print(f"Running {name}")
    try:
        Y = method.fit_transform(df.to_numpy())
        ax.scatter(Y[:, 0], Y[:, 1], c=target)
    except Exception:
        print(f"Method {name} failed")
    ax.set_title(name)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
```

# User Defined Functions

```python
data2 = np.random.normal(10, 2, size=(100000, 2))
df2 = pd.DataFrame(data2, columns=["x", "y"])
df2
```

```python
hypot = (df2.x**2 + df2.y**2)**0.5
print(hypot[0])
```

```python
# Not recommmended
def hypot1(x, y):
    return np.sqrt(x**2 + y**2)

h1 = []
for index, (x, y) in df2.iterrows():
    h1.append(hypot1(x, y))
print(h1[0])
```

```python
# Not recommended
def hypot2(row):
    return np.sqrt(row.x**2 + row.y**2)

h2 = df2.apply(hypot2, axis=1)
print(h2[0])
```

# Grouping - Aggregation

```python
import pandas as pd
import numpy as np

df = pd.read_csv("train.csv", low_memory=False, parse_dates=["Date"])  
df.head()
```

# Single aggregates

```python
df.groupby("Store").mean().head()
```

# Different Aggregates for Different Columns

```python
df.groupby(["Store", "DayOfWeek"]).agg({"Sales": "mean", "Customers": "count"})
```

```python
df2 = df.groupby(["Store", "DayOfWeek"]).agg({"Sales": ["mean", "max", "min"], "Customers": "count"})
df2.head()
```

```python
df2.columns = ["SalesMean", "SalesMax", "SalesMin", "Count"]
df2.head()
```

```python
# not recommended
mc_uncert = lambda x: np.std(x) / np.sqrt(x.size)
df2 = df.groupby(["Store", "DayOfWeek"]).agg(
    {"Sales": [("SalesMean", "mean"), ("SalesUncert", mc_uncert)], "Customers": "count"})
df2.head()
```

```python
# recommended
def mc_uncert2(x):
    return np.std(x) / np.sqrt(x.size)

dfg = df.groupby(["Store", "DayOfWeek"])
dfg.agg(
    SalesMean=("Sales", "meaicln"),
    SalesUncert=("Sales", mc_uncert2)
).reset_index().head()
```
